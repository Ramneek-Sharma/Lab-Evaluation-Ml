```python
import pandas as pd

```

```python
df = pd.read_csv("Fuel_cell_performance_data-Full.csv")
```

```python
target_column = 'Target1'
df = df.drop(columns=['Target2', 'Target3', 'Target4', 'Target5'], errors='ignore')
```

```python
exp = setup(data=df, target=target_column, train_size=0.7, session_id=42, verbose=False)
```

```python
best_model = compare_models()
```

```python
print("\nBest Model Identified:")
print(best_model)
```

```python
save_model(best_model, 'best_regression_model')
```

```python
predictions = predict_model(best_model)
print(predictions)
```
